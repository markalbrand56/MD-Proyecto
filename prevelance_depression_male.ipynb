# Resumen variables numéricas
## Prevelance_depression_male

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# ignore all FutureWarnings
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
# con pandas, leer une xlsx
excel_file = pd.ExcelFile('Mental health Depression disorder Data (1).xlsx')

# Sheet prevalence-by-mental-and-substa
dfs = {}  # Diccionario para almacenar los DataFrames
for sheet_name in excel_file.sheet_names:
    dfs[sheet_name] = excel_file.parse(sheet_name)
```

```python
# Sheet prevalence-of-depression-males-
prevalence_depression = dfs['prevalence-of-depression-males-']
```

```python
prevalence_depression = prevalence_depression.dropna()
prevalence_depression.head()
```

### Variables númericas

```python
prevalence_depression.dtypes
```

| Nombre                    | Tipo de Dato | Tipo de variable      |
|---------------------------|--------------|-----------------------|
| Entity                    | object       | Cualitativa Nominal   |
| Code                      | object       | Cualitativa Nominal   |
| Year                      | int64        | Cuantitativa Discreta |
| Prevalence in males (%)   | float64      | Cuantitativa Continua |
| Prevalence in females (%) | float64      | Cuantitativa Continua |
| Population      | float64      | Cuantitativa Continua |


### Análisis de variables cuantitativas

```python
# generar caja de bigotes para Prevalence in
plt.figure(figsize=(10, 6))
sns.boxplot(data=prevalence_depression[['Prevalence in males (%)', 'Prevalence in females (%)']])
plt.xticks(rotation=90)
plt.show()
```

```python
# generar caja de bigotes para Population
plt.figure(figsize=(10, 6))
sns.boxplot(data=prevalence_depression[['Population']])
plt.xticks(rotation=90)
plt.show()
```

De las variables analizadas, dos de estas (prevalencia en hombres, prevalencia en mujeres) evalúan en %, por lo que se puede decir que estas están en la misma escala.
Ahora la población es el número total de cada país.
Con los diagramas anteriores se puede observar que la variable con menos dispersión es prevalencia en hombres, mientras que la variable con mayor dispersión es la de la población.

### Análisis de variables cuantitativas

#### Histogramas

##### Año
Tipo de distribución: Uniforme


```python
# suicide rate
plt.hist(prevalence_depression['Year'], bins=10)
plt.title('Histograma de Año')
plt.show()
```

##### Prevalencia en hombres
Tipo de distribución: Asimétrica Positiva

```python
# prevalence male
plt.hist(prevalence_depression['Prevalence in males (%)'], bins=10)
plt.title('Histograma de la prevalencia en hombres')
plt.show()
```

##### Prevalencia en mujeres
Tipo de distribución: Asimétrica Positiva

```python
# prevalence male
plt.hist(prevalence_depression['Prevalence in females (%)'], bins=10)
plt.title('Histograma de la prevalencia en females')
plt.show()
```

##### Población
Tipo de distribución: J' Invertida

```python
# poblacion
plt.hist(prevalence_depression['Population'], bins=10)
plt.title('Histograma de Poblacion')
plt.show()
```

#### Análisis de correlación
Dentro de este set de datos, se puede observar que las variables númericas con mayor correlación son:
- La prevalencia de hombre y la prevalencia de mujeres

```python
matriz_correlacion_pms = prevalence_depression[['Prevalence in males (%)', 'Prevalence in females (%)', 'Population']].corr()

matriz_correlacion_pms
```

```python
# Pairplot
sns.pairplot( prevalence_depression[['Prevalence in males (%)', 'Prevalence in females (%)', 'Population']])
plt.title('Pairplot')
plt.show()
```

```python
# Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(matriz_correlacion_pms, annot=True)
plt.title('Mapa de Calor')
plt.show()
```
